## ALL filter functions

```python


# Helper functions
def lerp(a, b, l):
    """Linear interpolation between two values."""
    return a + l * (b - a)

def rgb2lum(image):
    """Convert RGB image to luminance."""
    r, g, b = image[..., 0], image[..., 1], image[..., 2]
    return 0.27 * r + 0.67 * g + 0.06 * b

# White Balance Filter
class WhiteBalanceFilter:
    def __init__(self, param_range=(0.7, 1.3)):
        self.param_range = param_range

    def process(self, img, params):
        """Apply white balance transformation to the image."""
        # Reshape params to [1, 1, 3] for element-wise multiplication
        params_reshaped = np.reshape(params, [1, 1, 3])
        return img * params_reshaped

    def random_params(self):
        """Generate random white balance parameters for RGB channels."""
        return [random.uniform(self.param_range[0], self.param_range[1]) for _ in range(3)]

# Gamma Filter
class GammaFilter:
    def __init__(self, gamma_range=(0.5, 3.0)):
        self.gamma_range = gamma_range

    def process(self, img, gamma_param):
        """Apply gamma correction to the image."""
        # Prevent numerical issues with very small values
        img_safe = np.maximum(img, 1e-4)
        # Apply gamma correction
        return np.power(img_safe, gamma_param)

    def random_param(self):
        """Generate a random gamma parameter."""
        return random.uniform(self.gamma_range[0], self.gamma_range[1])

# Tone Filter
class ToneFilter:
    def __init__(self, curve_steps=8, tone_range=(0.5, 2.0)):
        self.curve_steps = curve_steps
        self.tone_range = tone_range

    def process(self, img, params):
        """Apply tone adjustments to the image."""
        # Initialize with zeros
        total_image = np.zeros_like(img)
        tone_curve_sum = np.sum(params) + 1e-30

        # Apply tone adjustments step by step
        for i in range(self.curve_steps):
            step_contribution = np.clip(img - 1.0 * i / self.curve_steps, 0, 1.0 / self.curve_steps) * params[i]
            total_image += step_contribution

        # Normalize by the sum of parameters
        total_image *= self.curve_steps / tone_curve_sum
        return total_image

    def random_params(self):
        """Generate random tone adjustment parameters."""
        return [random.uniform(self.tone_range[0], self.tone_range[1]) for _ in range(self.curve_steps)]

# Contrast Filter
class ContrastFilter:
    def __init__(self, contrast_range=(0.1, 1.0)):
        self.contrast_range = contrast_range

    def process(self, img, contrast_param):
        """Apply contrast adjustment to the image."""
        # Calculate luminance
        lum = rgb2lum(img)
        lum = np.clip(lum, 0.0, 1.0)

        # Create contrast based on luminance
        contrast_lum = -np.cos(np.pi * lum) * 0.5 + 0.5

        # Expand dimensions to match the original image
        lum_expanded = np.expand_dims(lum, axis=-1)
        contrast_lum_expanded = np.expand_dims(contrast_lum, axis=-1)

        # Calculate contrast image
        contrast_image = img / (lum_expanded + 1e-6) * contrast_lum_expanded

        # Blend original and contrast image
        return lerp(img, contrast_image, contrast_param)

    def random_param(self):
        """Generate a random contrast parameter."""
        return random.uniform(self.contrast_range[0], self.contrast_range[1])

# USM (Unsharp Mask) Filter for sharpening
class UsmFilter:
    def __init__(self, usm_range=(0.1, 1.0)):
        self.usm_range = usm_range
        self.kernel = np.array([[0, -1, 0],
                                [-1, 5, -1],
                                [0, -1, 0]], dtype=np.float32)

    def process(self, img, usm_param):
        """Apply unsharp mask filter to sharpen the image."""
        # Apply sharpening kernel to each channel
        sharpened = np.zeros_like(img)
        for c in range(3):
            sharpened[:, :, c] = cv2.filter2D(img[:, :, c], -1, self.kernel)

        # Blend original and sharpened image
        return lerp(img, sharpened, usm_param)

    def random_param(self):
        """Generate a random USM parameter."""
        return random.uniform(self.usm_range[0], self.usm_range[1])

# Fog Filter
class FogFilter:
    def __init__(self, param_range=(0.3, 1.0)):
        self.param_range = param_range

    def synthetic_IcA_map(self, shape, noise_level=0.1, uniform=False, constant_value=1.0):
        """Create a synthetic IcA map."""
        H, W = shape[0], shape[1]

        if uniform:
            IcA_map = np.full((H, W), constant_value, dtype=np.float32)
        else:
            # Create a vertical gradient: 0 at the top, 1 at the bottom
            gradient = np.linspace(0, 1, H).reshape(H, 1)
            gradient = np.tile(gradient, (1, W))
            # Add uniform noise
            noise = np.random.uniform(0, noise_level, (H, W))
            IcA_map = np.clip(gradient + noise, 0, 1)

        return IcA_map

    def process(self, img, defog_param, use_uniform_fog=False):
        """Apply fog effect to the image using atmospheric scattering model."""
        # Create IcA map
        IcA = self.synthetic_IcA_map(img.shape[:2], noise_level=0.1,
                                     uniform=use_uniform_fog, constant_value=1.0)

        # Fix atmospheric light to light gray
        A = np.array([0.9, 0.9, 0.9], dtype=np.float32)

        # Compute transmission map
        t = 1 - defog_param * IcA
        t = np.clip(t, 0.01, 1.0)

        # Reshape for broadcasting
        A_reshaped = A.reshape((1, 1, 3))
        t_expanded = np.expand_dims(t, axis=-1)

        # Apply atmospheric scattering model: I(x) = t(x)*J(x) + (1-t(x))*A
        foggy_image = t_expanded * img + (1 - t_expanded) * A_reshaped

        return foggy_image

    def random_params(self):
        """Generate random fog parameters."""
        defog_param = random.uniform(self.param_range[0], self.param_range[1])
        use_uniform_fog = True if random.random() < 0.35 else False
        return defog_param, use_uniform_fog

# Main Pipeline Class
class ImageFilterPipeline:
    def __init__(self):
        self.wb_filter = WhiteBalanceFilter()
        self.gamma_filter = GammaFilter()
        self.tone_filter = ToneFilter()
        self.contrast_filter = ContrastFilter()
        self.usm_filter = UsmFilter()
        self.fog_filter = FogFilter()

    def process_image(self, img):
        """Apply all filters with random parameters and return the processed image and parameters."""
        # Initialize parameters dictionary
        params = {}

        # Apply White Balance
        wb_params = self.wb_filter.random_params()
        img = self.wb_filter.process(img, wb_params)
        params["wb_r"], params["wb_g"], params["wb_b"] = wb_params

        # Apply Gamma Correction
        gamma_param = self.gamma_filter.random_param()
        img = self.gamma_filter.process(img, gamma_param)
        params["gamma"] = gamma_param

        # Apply Tone Adjustment
        tone_params = self.tone_filter.random_params()
        img = self.tone_filter.process(img, tone_params)
        for i, param in enumerate(tone_params):
            params[f"tone_{i}"] = param

        # Apply Contrast
        contrast_param = self.contrast_filter.random_param()
        img = self.contrast_filter.process(img, contrast_param)
        params["contrast"] = contrast_param

        # Apply USM Sharpening
        usm_param = self.usm_filter.random_param()
        img = self.usm_filter.process(img, usm_param)
        params["usm"] = usm_param

        # Apply Fog Effect
        defog_param, use_uniform_fog = self.fog_filter.random_params()
        img = self.fog_filter.process(img, defog_param, use_uniform_fog)
        params["fog_param"] = defog_param
        params["uniform_fog"] = 1 if use_uniform_fog else 0

        # Clip values to valid range
        img = np.clip(img, 0.0, 1.0)

        return img, params

def process_single_image(img_path, output_dir, pipeline):
    """Process a single image and return its parameters."""
    try:
        # Extract filename without extension
        filename = os.path.basename(img_path)
        base_filename, ext = os.path.splitext(filename)
        output_filename = f"synthetic_{base_filename}{ext}"
        output_path = os.path.join(output_dir, output_filename)

        # Read the image with reduced size to save memory
        img_bgr = cv2.imread(img_path)
        if img_bgr is None:
            return None


        max_size = 1024  # Maximum dimension (width or height)
        h, w = img_bgr.shape[:2]
        if max(h, w) > max_size:
            scale = max_size / max(h, w)
            new_h, new_w = int(h * scale), int(w * scale)
            img_bgr = cv2.resize(img_bgr, (new_w, new_h), interpolation=cv2.INTER_AREA)

        # Convert BGR to RGB and normalize to [0, 1]
        img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
        img_rgb = img_rgb.astype(np.float32) / 255.0

        # Apply the filter pipeline
        processed_img, params = pipeline.process_image(img_rgb)

        # Add filename to parameters
        params["filename"] = output_filename

        # Convert back to uint8 and BGR for saving
        processed_img_uint8 = (processed_img * 255).astype(np.uint8)
        processed_img_bgr = cv2.cvtColor(processed_img_uint8, cv2.COLOR_RGB2BGR)

        # Save the processed image
        cv2.imwrite(output_path, processed_img_bgr)

        # Clear memory
        del img_bgr, img_rgb, processed_img, processed_img_uint8, processed_img_bgr
        gc.collect()

        return params
    except Exception as e:
        print(f"Error processing {img_path}: {str(e)}")
        return None

def batch_process_directory(input_dir, output_dir, csv_path, batch_size=100, max_images=None, num_workers=None):
    """
    Process images in batches to manage memory usage.

    Args:
        input_dir (str): Directory containing input images
        output_dir (str): Directory to save processed images
        csv_path (str): Path to save parameter CSV file
        batch_size (int): Number of images to process in each batch
        max_images (int, optional): Maximum number of images to process
        num_workers (int, optional): Number of worker processes to use
    """
    # Create output directory if it doesn't exist
    os.makedirs(output_dir, exist_ok=True)

    # Set number of workers
    if num_workers is None:
        num_workers = max(1, multiprocessing.cpu_count() - 1)

    # Get all image files in the input directory
    extensions = ["*.jpg", "*.jpeg", "*.png"]
    image_paths = []
    for ext in extensions:
        image_paths.extend(glob(os.path.join(input_dir, ext)))

    if not image_paths:
        raise ValueError(f"No images found in {input_dir}")

    # If max_images specified, limit the number of images
    if max_images is not None:
        image_paths = image_paths[:max_images]

    # Initialize the pipeline once
    pipeline = ImageFilterPipeline()

    # Process images in batches
    all_params = []
    total_images = len(image_paths)
    num_batches = (total_images + batch_size - 1) // batch_size

    # Function to write CSV in chunks
    def write_params_to_csv(params_list, csv_path, mode='a'):
        if not params_list:
            return

        # Convert params to DataFrame
        df = pd.DataFrame(params_list)

        # Check if file exists for header
        file_exists = os.path.isfile(csv_path)

        # Write to CSV
        df.to_csv(csv_path, mode=mode, index=False, header=not file_exists or mode=='w')

    # Clear CSV file if it exists
    if os.path.exists(csv_path):
        os.remove(csv_path)

    print(f"Processing {total_images} images in {num_batches} batches with {num_workers} workers")

    for batch_idx in range(num_batches):
        start_idx = batch_idx * batch_size
        end_idx = min(start_idx + batch_size, total_images)
        current_batch = image_paths[start_idx:end_idx]

        print(f"Processing batch {batch_idx+1}/{num_batches} ({len(current_batch)} images)")

        # Process batch in parallel
        process_func = partial(process_single_image, output_dir=output_dir, pipeline=pipeline)

        batch_params = []
        if num_workers > 1:
            with multiprocessing.Pool(processes=num_workers) as pool:
                batch_params = list(tqdm(pool.imap(process_func, current_batch),
                                        total=len(current_batch),
                                        desc=f"Batch {batch_idx+1}/{num_batches}"))
        else:
            # Sequential processing if multiprocessing is disabled
            batch_params = [process_func(img_path) for img_path in tqdm(current_batch,
                                                                       desc=f"Batch {batch_idx+1}/{num_batches}")]

        # Filter out None values (failed processing)
        batch_params = [p for p in batch_params if p is not None]

        # Write this batch to CSV
        write_params_to_csv(batch_params, csv_path, mode='a')

        # Clear memory
        del batch_params
        gc.collect()

    print(f"Processing complete. Processed images saved to: {output_dir}")
    print(f"Parameters saved to: {csv_path}")
```

```python


def copy_file(args):
    """Helper function for parallel copying"""
    src_file, dest_file = args
    try:
        shutil.copy2(src_file, dest_file)
        return True
    except Exception as e:
        return False

def copy_random_images(source_folder, destination_folder, num_images=3500, csv_file=None):
    """
    Copy random images from source folder to destination folder with progress bar
    and optimized for systems with limited resources

    Args:
        source_folder: Path to the source folder containing images
        destination_folder: Path to the destination folder
        num_images: Number of random images to copy (default: 3500)
        csv_file: Optional CSV file to copy to the destination folder
    """
    # Create destination folder if it doesn't exist
    os.makedirs(destination_folder, exist_ok=True)

    print("Scanning for image files...")

    # Common image extensions
    image_extensions = {'.jpg', '.jpeg', '.png', '.gif', '.bmp', '.tiff'}

    # Use set for faster lookup
    all_files = []

    # More efficient file collection using list comprehension
    for root, _, files in os.walk(source_folder):
        all_files.extend(
            os.path.join(root, file) for file in files
            if os.path.splitext(file)[1].lower() in image_extensions
        )

    print(f"Found {len(all_files)} images in the source folder")

    # Check if we have enough images
    if len(all_files) < num_images:
        print(f"Warning: Only {len(all_files)} images available, which is less than requested {num_images}")
        selected_files = all_files
    else:
        # Select random images without replacement
        # Using random.sample which is more efficient than shuffling the whole list
        selected_files = random.sample(all_files, num_images)

    # Pre-check for duplicate filenames to avoid checks during copying
    filenames = {}
    copy_tasks = []

    for i, src_path in enumerate(selected_files):
        filename = os.path.basename(src_path)

        # Handle duplicate filenames
        if filename in filenames:
            name, ext = os.path.splitext(filename)
            filename = f"{name}_{i}{ext}"

        filenames[filename] = True
        dest_path = os.path.join(destination_folder, filename)
        copy_tasks.append((src_path, dest_path))

    # Determine optimal number of workers based on CPU cores, but limit for RAM constraints
    max_workers = min(multiprocessing.cpu_count(), 4)  # Limiting to 4 due to RAM constraints

    # Batch processing to limit memory usage
    batch_size = 1500  # Adjust based on RAM availability

    total_copied = 0

    for i in range(0, len(copy_tasks), batch_size):
        batch = copy_tasks[i:i+batch_size]

        with tqdm(total=len(batch), desc=f"Copying batch {i//batch_size + 1}/{(len(copy_tasks)+batch_size-1)//batch_size}") as pbar:
            with concurrent.futures.ThreadPoolExecutor(max_workers=max_workers) as executor:
                # Use map to process files in parallel with progress bar
                for result in executor.map(copy_file, batch):
                    if result:
                        total_copied += 1
                    pbar.update(1)

    print(f"Successfully copied {total_copied} images to {destination_folder}")

    # Copy the CSV file if specified
    if csv_file and os.path.exists(csv_file):
        csv_filename = os.path.basename(csv_file)
        csv_dest_path = os.path.join(destination_folder, csv_filename)
        try:
            shutil.copy2(csv_file, csv_dest_path)
            print(f"Copied CSV file to {csv_dest_path}")
        except Exception as e:
            print(f"Failed to copy CSV file: {e}")
    else:
        print("CSV file not found or not specified")

# Usage example
source_folder = "/content/drive/MyDrive/Info_assurance/synthetic_data/train"  # Replace with your source folder
destination_folder = "/content/drive/MyDrive/Info_assurance/synthetic_data/train_sample"  # Replace with your destination folder
csv_file = "/content/drive/MyDrive/Info_assurance/synthetic_data/train/train_synthetic_params.csv"

# Execute the function
copy_random_images(source_folder, destination_folder, num_images=20000, csv_file=csv_file)
```

```python
import os
import pandas as pd
from tqdm.notebook import tqdm
import re

def create_matched_csv_for_images(image_folder, csv_path, output_csv_path):
    """
    Create a new CSV file that contains exactly one row for each image in the folder,
    matching on the 'filename' column in the CSV.

    Args:
        image_folder: Path to the folder containing the images
        csv_path: Path to the original CSV file
        output_csv_path: Path where the matched CSV will be saved
    """
    # Get list of image files in the folder
    print("Scanning for image files...")
    image_files = [
        f for f in os.listdir(image_folder)
        if os.path.isfile(os.path.join(image_folder, f)) and
        any(f.lower().endswith(ext) for ext in ['.jpg', '.jpeg', '.png', '.gif', '.bmp', '.tiff'])
    ]

    print(f"Found {len(image_files)} images in the folder")

    # Function to extract the base filename without extension and any added suffixes
    def extract_base_name(filename):
        # Remove extension
        base = os.path.splitext(filename)[0]
        # Remove any added suffixes like _123 that might have been added during copying
        base = re.sub(r'_\d+$', '', base)
        return base

    # Process image filenames to extract base names
    print("Processing image filenames...")
    image_base_names = [extract_base_name(f) for f in image_files]

    # Load the CSV file
    print("Loading CSV file...")
    df = pd.read_csv(csv_path)

    # Verify the 'filename' column exists
    if 'filename' not in df.columns:
        print("Error: CSV does not contain a 'filename' column.")
        print("Available columns:", df.columns.tolist())
        return 0

    # Create a mapping of base names from CSV to row indices
    print("Creating mapping from CSV filenames to rows...")
    csv_name_to_row = {}

    for i, name in enumerate(df['filename']):
        if isinstance(name, str):
            base_name = extract_base_name(name)
            csv_name_to_row[base_name] = i
        else:
            # Handle non-string entries if they exist
            csv_name_to_row[str(name)] = i

    # Match each image to a CSV row
    print("Matching images with CSV entries...")
    matched_rows = []
    unmatched_images = []

    with tqdm(total=len(image_files), desc="Matching images to CSV rows") as pbar:
        for img_name, img_base_name in zip(image_files, image_base_names):
            if img_base_name in csv_name_to_row:
                # Found a direct match
                matched_rows.append(csv_name_to_row[img_base_name])
            else:
                # No match found for this image
                unmatched_images.append(img_name)
            pbar.update(1)

    print(f"Matched {len(matched_rows)} images to CSV rows")
    print(f"Unable to match {len(unmatched_images)} images")

    if unmatched_images:
        print(f"First 5 unmatched images: {unmatched_images[:5]}")

    # Create a new DataFrame with only the matched rows
    matched_df = df.iloc[matched_rows].copy()

    # Save the matched CSV
    print(f"Saving matched CSV with {len(matched_df)} rows...")
    matched_df.to_csv(output_csv_path, index=False)
    print(f"Matched CSV saved to {output_csv_path}")

    return len(matched_df)

# Example usage
image_folder = "/content/drive/MyDrive/Info_assurance/synthetic_data/train_sample"
original_csv = "/content/drive/MyDrive/Info_assurance/synthetic_data/train/train_synthetic_params.csv"
output_csv = "/content/drive/MyDrive/Info_assurance/synthetic_data/train_sample/train_synthetic_params_matched.csv"

# Run the function
matched_rows = create_matched_csv_for_images(image_folder, original_csv, output_csv)
print(f"Successfully created matched CSV with {matched_rows} entries matching image files")
```

```python


def copy_file(args):
    """Helper function for parallel copying"""
    src_file, dest_file = args
    try:
        shutil.copy2(src_file, dest_file)
        return True
    except Exception as e:
        return False

def copy_random_images(source_folder, destination_folder, num_images=3500, csv_file=None):
    """
    Copy random images from source folder to destination folder with progress bar
    and optimized for systems with limited resources

    Args:
        source_folder: Path to the source folder containing images
        destination_folder: Path to the destination folder
        num_images: Number of random images to copy (default: 3500)
        csv_file: Optional CSV file to copy to the destination folder
    """
    # Create destination folder if it doesn't exist
    os.makedirs(destination_folder, exist_ok=True)

    print("Scanning for image files...")

    # Common image extensions
    image_extensions = {'.jpg', '.jpeg', '.png', '.gif', '.bmp', '.tiff'}

    # Use set for faster lookup
    all_files = []

    # More efficient file collection using list comprehension
    for root, _, files in os.walk(source_folder):
        all_files.extend(
            os.path.join(root, file) for file in files
            if os.path.splitext(file)[1].lower() in image_extensions
        )

    print(f"Found {len(all_files)} images in the source folder")

    # Check if we have enough images
    if len(all_files) < num_images:
        print(f"Warning: Only {len(all_files)} images available, which is less than requested {num_images}")
        selected_files = all_files
    else:
        # Select random images without replacement
        # Using random.sample which is more efficient than shuffling the whole list
        selected_files = random.sample(all_files, num_images)

    # Pre-check for duplicate filenames to avoid checks during copying
    filenames = {}
    copy_tasks = []

    for i, src_path in enumerate(selected_files):
        filename = os.path.basename(src_path)

        # Handle duplicate filenames
        if filename in filenames:
            name, ext = os.path.splitext(filename)
            filename = f"{name}_{i}{ext}"

        filenames[filename] = True
        dest_path = os.path.join(destination_folder, filename)
        copy_tasks.append((src_path, dest_path))

    # Determine optimal number of workers based on CPU cores, but limit for RAM constraints
    max_workers = min(multiprocessing.cpu_count(), 4)  # Limiting to 4 due to RAM constraints

    # Batch processing to limit memory usage
    batch_size = 1500  # Adjust based on RAM availability

    total_copied = 0

    for i in range(0, len(copy_tasks), batch_size):
        batch = copy_tasks[i:i+batch_size]

        with tqdm(total=len(batch), desc=f"Copying batch {i//batch_size + 1}/{(len(copy_tasks)+batch_size-1)//batch_size}") as pbar:
            with concurrent.futures.ThreadPoolExecutor(max_workers=max_workers) as executor:
                # Use map to process files in parallel with progress bar
                for result in executor.map(copy_file, batch):
                    if result:
                        total_copied += 1
                    pbar.update(1)

    print(f"Successfully copied {total_copied} images to {destination_folder}")

    # Copy the CSV file if specified
    if csv_file and os.path.exists(csv_file):
        csv_filename = os.path.basename(csv_file)
        csv_dest_path = os.path.join(destination_folder, csv_filename)
        try:
            shutil.copy2(csv_file, csv_dest_path)
            print(f"Copied CSV file to {csv_dest_path}")
        except Exception as e:
            print(f"Failed to copy CSV file: {e}")
    else:
        print("CSV file not found or not specified")

# Usage example
source_folder = "/content/drive/MyDrive/Info_assurance/synthetic_data/train"  # Replace with your source folder
destination_folder = "/content/drive/MyDrive/Info_assurance/synthetic_data/train_sample2"  # Replace with your destination folder
csv_file = "/content/drive/MyDrive/Info_assurance/synthetic_data/train/train_synthetic_params.csv"

# Execute the function
copy_random_images(source_folder, destination_folder, num_images=15000, csv_file=csv_file)
```

```python
import os
import pandas as pd
from tqdm.notebook import tqdm
import re

def create_matched_csv_for_images(image_folder, csv_path, output_csv_path):
    """
    Create a new CSV file that contains exactly one row for each image in the folder,
    matching on the 'filename' column in the CSV.

    Args:
        image_folder: Path to the folder containing the images
        csv_path: Path to the original CSV file
        output_csv_path: Path where the matched CSV will be saved
    """
    # Get list of image files in the folder
    print("Scanning for image files...")
    image_files = [
        f for f in os.listdir(image_folder)
        if os.path.isfile(os.path.join(image_folder, f)) and
        any(f.lower().endswith(ext) for ext in ['.jpg', '.jpeg', '.png', '.gif', '.bmp', '.tiff'])
    ]

    print(f"Found {len(image_files)} images in the folder")

    # Function to extract the base filename without extension and any added suffixes
    def extract_base_name(filename):
        # Remove extension
        base = os.path.splitext(filename)[0]
        # Remove any added suffixes like _123 that might have been added during copying
        base = re.sub(r'_\d+$', '', base)
        return base

    # Process image filenames to extract base names
    print("Processing image filenames...")
    image_base_names = [extract_base_name(f) for f in image_files]

    # Load the CSV file
    print("Loading CSV file...")
    df = pd.read_csv(csv_path)

    # Verify the 'filename' column exists
    if 'filename' not in df.columns:
        print("Error: CSV does not contain a 'filename' column.")
        print("Available columns:", df.columns.tolist())
        return 0

    # Create a mapping of base names from CSV to row indices
    print("Creating mapping from CSV filenames to rows...")
    csv_name_to_row = {}

    for i, name in enumerate(df['filename']):
        if isinstance(name, str):
            base_name = extract_base_name(name)
            csv_name_to_row[base_name] = i
        else:
            # Handle non-string entries if they exist
            csv_name_to_row[str(name)] = i

    # Match each image to a CSV row
    print("Matching images with CSV entries...")
    matched_rows = []
    unmatched_images = []

    with tqdm(total=len(image_files), desc="Matching images to CSV rows") as pbar:
        for img_name, img_base_name in zip(image_files, image_base_names):
            if img_base_name in csv_name_to_row:
                # Found a direct match
                matched_rows.append(csv_name_to_row[img_base_name])
            else:
                # No match found for this image
                unmatched_images.append(img_name)
            pbar.update(1)

    print(f"Matched {len(matched_rows)} images to CSV rows")
    print(f"Unable to match {len(unmatched_images)} images")

    if unmatched_images:
        print(f"First 5 unmatched images: {unmatched_images[:5]}")

    # Create a new DataFrame with only the matched rows
    matched_df = df.iloc[matched_rows].copy()

    # Save the matched CSV
    print(f"Saving matched CSV with {len(matched_df)} rows...")
    matched_df.to_csv(output_csv_path, index=False)
    print(f"Matched CSV saved to {output_csv_path}")

    return len(matched_df)

# Example usage
image_folder = "/content/drive/MyDrive/Info_assurance/synthetic_data/train_sample2"
original_csv = "/content/drive/MyDrive/Info_assurance/synthetic_data/train/train_synthetic_params.csv"
output_csv = "/content/drive/MyDrive/Info_assurance/synthetic_data/train_sample/train_synthetic_params_matched.csv"

# Run the function
matched_rows = create_matched_csv_for_images(image_folder, original_csv, output_csv)
print(f"Successfully created matched CSV with {matched_rows} entries matching image files")
```
